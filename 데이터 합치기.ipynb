# 실습 Data (Kaggle : Life Expectancy)

```python
import pandas as pd

df_men = pd.read_csv('data/men-survival-to-age-65.csv')
df_women = pd.read_csv('data/women-survival-to-age-65.csv')
```

```python
df_men
```

```python
df_women
```

# merge

```python
pd.merge(df_men, df_women, on=['Entity','Year'])
```

```python
pd.merge(df_men, df_women, on=['Entity','Year'], suffixes=['_men', '_women'])
```

```python
df_corr = pd.merge(df_men, df_women, on=['Entity','Year'], suffixes=['_men', '_women'])
df_corr.corr()
```

# concat

```python
pd.concat([df_men, df_women], axis=0)
```

```python
df_men['sex'] = 'male'
df_women['sex'] = 'female'

df_men.columns  = ['Entity', 'Code', 'Year', 'Survival to age 65 (% of cohort)', 'sex']
df_women.columns  = ['Entity', 'Code', 'Year', 'Survival to age 65 (% of cohort)', 'sex']

pd.concat([df_men, df_women], axis=0)
```

```python
pd.concat([df_men, df_women], axis=1)
```
