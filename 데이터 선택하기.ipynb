# quantile (IQR)

```python
Q1 = df['Founded'].quantile(0.25)
Q3 = df['Founded'].quantile(0.75)
IQR = Q3 - Q1

df_after_IQR = df[(df['Founded'] <= Q3 + 1.5 * IQR) & (df['Founded'] >= Q1 - 1.5 * IQR )]
df_after_IQR
```
