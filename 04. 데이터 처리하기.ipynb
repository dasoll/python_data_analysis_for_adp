# stack

```python
df_men = pd.read_csv('data/men-survival-to-age-65.csv')
df_women = pd.read_csv('data/women-survival-to-age-65.csv')

df_merge = pd.merge(df_men, df_women, on=['Entity','Year'], suffixes=['_men', '_women'])

df_merge.corr()
```

```python
df_merge.corr().stack().reset_index()
```

```python
df_merge.corr().stack().reset_index().sort_values(by=0, ascending=False)
```

# apply

```python
df_merge = pd.merge(df_men, df_women, on=['Entity','Year'], suffixes=['_men', '_women'])
df_merge_numeric = df_merge[['Survival to age 65, male (% of cohort)', 'Survival to age 65, female (% of cohort)']]

df_merge_numeric.apply(np.mean)
```

```python
df_merge_numeric.apply(np.mean, axis=1)
```

```python
def make_category(sr):
    sr_new = sr.copy()
    
    for i, row in enumerate(sr_new):
        if row >= 50:
             sr_new[i] = '상'
        else:
            sr_new[i] = '하'
    
    return sr_new

df_category = df_merge_numeric.apply(make_category)
df_category
```

```python
df.groupby(['Entity'])['Survival to age 65, male (% of cohort)'].apply(np.round)
```

```python
def make_national_year_range(df):
    return df['Year'].max() - df['Year'].min()

df.groupby(['Entity']).apply(make_national_year_range)
```

# map

```python
def make_death_rate(value):
    return 100 - value

sr = df['Survival to age 65, male (% of cohort)']
sr.map(make_death_rate)
```

```python
df['death_rate'] = sr.map(make_death_rate)
df
```

```python
df_merge_numeric['Survival to age 65, male (% of cohort)'].map(lambda x: '상' if x>=50 else '하')
```
