# 실습 Data (서울 열린 데이터광장 : 서울시 소비자물가지수(지출목적별) 통계)

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('data/서울시 소비자물가지수(지출목적별) 통계.txt', delimiter='\t')
df = df.drop(0).reset_index(drop=True)
df
```

```python
# notebook에 보이는 최대 행과 최대 열의 개수를 늘려주는 설정

pd.set_option('display.max_row', 500)
pd.set_option('display.max_columns', 100)
```

```python
cpi = df[['기간','총지수']]
cpi.columns = ['date', 'value']
cpi
```

# to_datetime

```python
pd.to_datetime(cpi['date'])
```

```python
print(cpi['date'].dtype)
print(pd.to_datetime(cpi['date']).dtype)
```

```python
cpi['date'] = pd.to_datetime(cpi['date'])
```

# strptime, strftime

```python
import datetime as dt

dt.datetime.strptime('2022-06-25', '%Y-%m-%d')
```

```python
today = dt.datetime.today()

dt.datetime.strftime(today, '%Y.%m.%d')
```

```python
today.strftime('%Y/%m/%d')
```

```python
cpi['date'] = cpi['date'].dt.strftime('%Y.%m')
cpi['date']
```

```python
cpi['date'] = cpi['date'].map(lambda x: dt.datetime.strptime(x, '%Y.%m'))
cpi['date']
```

# date_range

```python
pd.date_range('2009-12', '2022-04', periods=None, freq='1M')
```

```python
sr_dt_range = pd.date_range('2009-12', '2022-04', freq='1M')
sr_dt_range = sr_dt_range.to_series().reset_index(drop=True)
sr_dt_range =  sr_dt_range + dt.timedelta(days=1)
```

```python
result = pd.concat([cpi['date'], sr_dt_range], axis=1)
result
```

```python
result.isnull().sum()
```

# shift

```python
cpi['value'] = cpi['value'].astype('float')
```

```python
cpi['value_year_ago'] = cpi['value'].shift(12)
cpi
```

```python
cpi['YoY'] = (cpi['value'] - cpi['value_year_ago']) / cpi['value_year_ago'] * 100
cpi.dropna().tail()
```

# rolling

```python
cpi['moving_average_3'] = cpi['value'].rolling(3).mean()
cpi.tail()
```

```python
cpi['value'].rolling(3).apply(lambda x: (x.iloc[0]*0.2 + x.iloc[1]*0.3 + x.iloc[2]*0.5))
```

```python
weight = np.array([0.2, 0.3, 0.5])

cpi['weight_moving_average_3'] = cpi['value'].rolling(3).apply(lambda x: weight.dot(x))
cpi.tail()
```
